```python
import sys
import io
import csv
from selenium import webdriver
from selenium.webdriver.common.by import By       
import time

# stored path
path_to_file = "/Users/kevin/Desktop/reviewdemoGrand_Hilai-Da_an_Taipei.csv"

# number of scraped pages
num_page = 694

# tripadvisor website of restaurant
url = "https://www.tripadvisor.com.tw/Restaurant_Review-g13811269-d8814379-Reviews-Grand_Hilai-Da_an_Taipei.html"


if (len(sys.argv) == 4):
    path_to_file = sys.argv[1]
    num_page = int(sys.argv[2])
    url = sys.argv[3]

# Import the webdriver
driver = webdriver.Chrome(executable_path='/Users/kevin/Web Clawer DEMO/chromedriver')
driver.get(url)

# open csv and save the review 
csvFile = open(path_to_file, 'a', encoding="utf-8")
csvWriter = csv.writer(csvFile)

# function to check if the button is on the page
from selenium.common.exceptions import NoSuchElementException        
def check_exists_by_xpath(xpath):
    try:
        driver.find_element_by_xpath(xpath)
    except NoSuchElementException:
        return False
    return True

# change the value of page inside the range to save more or less reviews
for i in range(0, num_page):
    
    if (check_exists_by_xpath("//span[@class='taLnk ulBlueLinks']")):
        # click more button and expand the review 
        driver.find_element_by_xpath("//span[@class='taLnk ulBlueLinks']").click()
        time.sleep(5)
    #driver.find_element(By.XPATH, "//span[@class='taLnk ulBlueLinks']").click()

    container = driver.find_elements_by_xpath(".//div[@class='review-container']")

    for j in range(len(container)):

        title = container[j].find_element_by_xpath(".//span[@class='noQuotes']").text
        date = container[j].find_element_by_xpath(".//span[contains(@class, 'ratingDate')]").get_attribute("title")
        rating = container[j].find_element_by_xpath(".//span[contains(@class, 'ui_bubble_rating bubble_')]").get_attribute("class").split("_")[3]
        review = container[j].find_element_by_xpath(".//p[@class='partial_entry']").text.replace("\n", " ")
        like = container[j].find_element_by_xpath(".//span[contains(@class, 'numHelp ')]").text
        csvWriter.writerow([date, title, review, rating, like]) 

    # change the page
    driver.find_element_by_xpath('.//a[@class="nav next ui_button primary"]').click()
    time.sleep(5)
driver.close()
```
